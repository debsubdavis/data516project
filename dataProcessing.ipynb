```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import regexp_replace

# Initialize Spark session
spark = SparkSession.builder \
    .appName("Local CSV Ingestion") \
    .config("spark.master", "local") \
    .getOrCreate()

# Read CSV file into DataFrame
df = spark.read.csv("/Users/debbiesubocz/uw-code/data516/project/output_csv/aka_name_202311192056.csv", header=True, inferSchema=True)

# Drop duplicates based on the ID column
df_clean = df.dropDuplicates(["id"])  # Replace 'id' with your actual column name for ID

# Clean pipe characters from each column
for column in df_clean.columns:
    df_clean = df_clean.withColumn(column, regexp_replace(column, "\\|", " "))

# Display the result
df_clean.show()

# Coalesce the DataFrame to a single partition, set the delimiter to |, and write to a single CSV file
df_clean.coalesce(1).write.option("delimiter", "|").csv("/Users/debbiesubocz/uw-code/data516/project/output_csv/cleaned_aka_name.csv", header=True, mode="overwrite")

# Note: mode="overwrite" will overwrite the directory if it already exists
```
